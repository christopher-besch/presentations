```python
from manim import *
```

```python
%%manim Tree

class Tree(Scene):
    def construct(self):
        vertices = range(11)
        labels = {0: "+",
                  1: "\cdot", 2: "\cdot",
                  3: "+", 4: "3", 5: "4", 6: "+",
                  7: "35", 8: "15", 9: "31", 10: "9"}
        edges = [
            (0, 1), (0, 2),
            (1, 3), (1, 4), (2, 5), (2, 6),
            (3, 7), (3, 8), (6, 9), (6, 10),
        ]
        g = Graph(vertices, edges, layout="tree", labels=labels, root_vertex=0, layout_scale=3)
        self.add(g)
```
